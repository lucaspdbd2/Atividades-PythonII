#### Q3. 
Crie um dataframe `participacao` com as colunas `país`, `copa`, e `num_copas` que armazena a quantidade de copas do mundo que cada país participou, informando se é da copa masculina ou feminina. Ao final imprima, como apresentado a seguir, o top 5 países de cada competição que mais participou de copas do mundo.

Exemplo de saída (valores inventados):
```
país            copa        num_copas
Brazil          Feminina    8
Unites States   Feminina    8
Germany         Feminina    8
Japan           Feminina    7
Colombia        Feminina    7
```

```
país            copa        num_copas
Brazil          Masculina   21
Germany         Masculina   21
Argentina       Masculina   20
England         Masculina   20
Mexico          Masculina   20
```

````python
# 1. Isolar os países das colunas time_1 e time_2 junto com ano e copa
times_1 = wc[['time_1', 'ano', 'copa']].rename(columns={'time_1': 'país'})
times_2 = wc[['time_2', 'ano', 'copa']].rename(columns={'time_2': 'país'})

# 2. Juntar tudo em um único dataframe
todas_participacoes = pd.concat([times_1, times_2])

# 3. Remover duplicatas (para que um país não seja contado várias vezes no mesmo ano)
participacoes_unicas = todas_participacoes.drop_duplicates()

# 4. Agrupar por país e copa, e contar a quantidade de anos
participacao = participacoes_unicas.groupby(['país', 'copa'])['ano'].count().reset_index()
participacao.rename(columns={'ano': 'num_copas'}, inplace=True)

# 5. Ordenar pela quantidade de participações (decrescente)
participacao = participacao.sort_values(by=['copa', 'num_copas'], ascending=[True, False])

# 6. Filtrar o Top 5 de cada categoria
top5_feminina = participacao[participacao['copa'] == 'Feminina'].head(5)
top5_masculina = participacao[participacao['copa'] == 'Masculina'].head(5)

# 7. Imprimir os resultados (usando to_string para remover o índice e igualar ao exemplo)
print(top5_feminina.to_string(index=False))
print("\n")
print(top5_masculina.to_string(index=False))
````

#### Q4. 
* Crie um dataframe `gols` com duas colunas `país` e `total_gols` com o total de gols marcados por cada país durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).
* Imprima o dataframe `gols` ordenado de forma decrescente, para que os times com mais gols fiquem no topo.

Segue um exemplo ilustrativo com o formato do dataframe resultado:

```
país        total_gols
Brazil      120
Argentina   112
Germany     107
...
```

````python
# 1. Separar os gols marcados como mandante (time_1) e visitante (time_2)
gols_mandante = wc[['time_1', 'gols_1']].rename(columns={'time_1': 'país', 'gols_1': 'gols'})
gols_visitante = wc[['time_2', 'gols_2']].rename(columns={'time_2': 'país', 'gols_2': 'gols'})

# 2. Juntar os dois em um único dataframe
todos_gols = pd.concat([gols_mandante, gols_visitante])

# 3. Agrupar pelo país e somar todos os gols
gols = todos_gols.groupby('país')['gols'].sum().reset_index()

# 4. Renomear a coluna de soma para bater com a exigência do enunciado
gols = gols.rename(columns={'gols': 'total_gols'})

# 5. Ordenar o dataframe pelo total de gols de forma decrescente
gols = gols.sort_values(by='total_gols', ascending=False)

# 6. Imprimir o resultado
print(gols.to_string(index=False))
````

#### Q5. 
Qual país tomou mais cartões amarelos somando todas as copas?

Neste exercício você vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observação é uma string represetando uma lista dos cartões amarelos de um único jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16’|1:0|Rosana Gómez', '20’|2:0|Gabriela Chávez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o método genérico `apply` para chamar uma função que você vai criar para transformar uma observação de cartão amarelo em uma contagem de elementos da lista. 

Lembre de levar em consideração que muitas observações são `NaN`. 

Em seguida faça sua mágica para agrupar as informações por país, acumular os cartões de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
país        cartões amarelos
Argentina   97
England     93
Australia   93
...
```

````python
# 1. Função para contar os cartões dentro da string
def contar_cartoes(valor):
    # Trata os valores nulos (NaN) retornando 0
    if pd.isna(valor):
        return 0
    
    # Converte para string, caso não seja, e remove os colchetes '[ ]' das pontas
    valor_limpo = str(valor).strip("[]")
    
    # Se, após remover os colchetes, não sobrar nada, significa 0 cartões
    if valor_limpo.strip() == "":
        return 0
        
    # Conta a quantidade de elementos dividindo o texto pelas vírgulas
    return len(valor_limpo.split(','))

# 2. Cria as novas colunas numéricas usando o método apply
wc['num_cartoes_1'] = wc['cartao_amarelo_1'].apply(contar_cartoes)
wc['num_cartoes_2'] = wc['cartao_amarelo_2'].apply(contar_cartoes)

# 3. Separa os cartões do time mandante (time_1) e do visitante (time_2)
cartoes_time1 = wc[['time_1', 'num_cartoes_1']].rename(columns={'time_1': 'país', 'num_cartoes_1': 'cartões amarelos'})
cartoes_time2 = wc[['time_2', 'num_cartoes_2']].rename(columns={'time_2': 'país', 'num_cartoes_2': 'cartões amarelos'})

# 4. Empilha (junta) os dois dataframes
todos_cartoes = pd.concat([cartoes_time1, cartoes_time2])

# 5. Agrupa os dados pelo país e soma o total de cartões, ordenando do maior para o menor
ranking_cartoes = todos_cartoes.groupby('país')['cartões amarelos'].sum().reset_index()
ranking_cartoes = ranking_cartoes.sort_values(by='cartões amarelos', ascending=False)

# 6. Exibe o resultado final com a formatação exigida (sem o índice numérico)
print(ranking_cartoes.to_string(index=False))
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de pênalti.

Para conseguir essa informação, você precisará trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas também são colunas textuais, onde cada observações apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol está na forma `'Jogador(a) · minuto’'`. Por exemplo:
```
'Alex Morgan · 12’|Rose Lavelle · 20’'
```

Lembre de levar em consideração que muitas observações são `NaN`. 

Recomendo criar um dicionário à parte, onde cada chave será um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente será a contagem de ocorrências desses nomes.

Em seguida basta converter o seu dicionário em um dataframe e imprimí-lo na forma (valores não são inventados 😁):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
# 1. Cria o dicionário vazio para armazenar a contagem
contagem_gols = {}

# 2. Cria uma função para extrair e contar os gols de uma única célula (observação)
def processar_gols(texto):
    # Ignora valores nulos (NaN)
    if pd.isna(texto):
        return
    
    # Converte para string e limpa colchetes ou aspas nas pontas (caso existam)
    texto = str(texto).strip("[]'")
    
    if not texto:
        return
        
    # Separa os múltiplos gols da partida pelo caractere pipe '|'
    gols = texto.split('|')
    
    for gol in gols:
        # Separa o nome do jogador do minuto do gol pelo caractere '·'
        partes = gol.split('·')
        
        if len(partes) > 0:
            # Pega a primeira parte (o nome), removendo espaços e aspas extras
            jogador = partes[0].strip(" '\"")
            
            # Se o jogador já tem gol computado, soma 1. Se não, inicia com 1.
            if jogador:
                if jogador in contagem_gols:
                    contagem_gols[jogador] += 1
                else:
                    contagem_gols[jogador] = 1

# 3. Aplica a função nas 4 colunas que contêm os detalhes dos gols
colunas_de_gols = ['gols_1_detalhes', 'gols_2_detalhes', 'gols_1_penalti', 'gols_2_penalti']

for coluna in colunas_de_gols:
    wc[coluna].apply(processar_gols)

# 4. Converte o dicionário populado em um novo DataFrame
df_artilheiros = pd.DataFrame(list(contagem_gols.items()), columns=['jogador(a)', 'num_gols'])

# 5. Ordena do maior para o menor e pega os 10 primeiros (top 10)
top_10_artilheiros = df_artilheiros.sort_values(by='num_gols', ascending=False).head(10)

# 6. Exibe o resultado final ocultando o índice, conforme o exemplo
print(top_10_artilheiros.to_string(index=False))
````
