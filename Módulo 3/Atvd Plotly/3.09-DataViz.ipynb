#### Q2.
Sobre os dados de copa do mundo, qual a distribuição de público presente nos jogos? Isso pode ser respondido com um histograma com os dados da coluna `comparecimento`.  

Lembre-se que alguns jogos estavam com público 0 incorretamente, que tal remover essas ocorrências para não atrapalhar sua visualzação?

Você deve implementar essa visualização nas três bibliotecas que vimos:
* Matplotlib
* Seaborn
* Plotly

Garanta que o gráfico tenha pelo menos os atributos de título e rótulos de dimensão.

````python
#### Solução com matplotlib
# Filtrando os dados para remover público igual a 0
wc_filtrado = wc[wc['comparecimento'] > 0]

# Criando o histograma com Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(wc_filtrado['comparecimento'], bins=30, edgecolor='black', alpha=0.7)

# Adicionando títulos e rótulos
plt.title('Distribuição de Público Presente nos Jogos da Copa do Mundo', fontsize=14)
plt.xlabel('Comparecimento (Público)')
plt.ylabel('Frequência (Quantidade de Jogos)')

# Exibindo o gráfico
plt.show()
````

````python
#### solução com seaborn
# Criando o histograma com Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=wc_filtrado, x='comparecimento', bins=30, kde=True)

# Adicionando títulos e rótulos usando a API do pyplot
plt.title('Distribuição de Público Presente nos Jogos da Copa do Mundo', fontsize=14)
plt.xlabel('Comparecimento (Público)')
plt.ylabel('Frequência (Quantidade de Jogos)')

# Exibindo o gráfico
plt.show()
````

````python
#### solução com plotly
# Criando o histograma interativo com Plotly Express
fig = px.histogram(
    wc_filtrado, 
    x='comparecimento', 
    nbins=30,
    title='Distribuição de Público Presente nos Jogos da Copa do Mundo',
    labels={'comparecimento': 'Comparecimento (Público)', 'count': 'Frequência (Quantidade de Jogos)'}
)

# Ajustando o layout para atualizar o nome do eixo Y (caso não mude automaticamente)
fig.update_layout(yaxis_title_text='Frequência (Quantidade de Jogos)')

# Exibindo o gráfico interativo
fig.show()
````

#### Q3.

Apresente um gráfico de dispersão (scatter) dos atributos `gols_1` e `gols_2`. Isso representa a relação entre gols feitos e gols tomados por jogo. Há alguma relação interessante entre esses atributos?

Para facilitar a visualização dos dados (já que tem muitos placares repetidos), aplique uma leve distorção aos dados para que cada ponto esteja deslocado aleatoriamente de seu valor original. Código apresentado a seguir
```python
gols = wc[['gols_1', 'gols_2']] * np.random.random((len(wc),2))
```

Você deve implementar essa visualização nas três bibliotecas que vimos:
* Matplotlib
* Seaborn
* Plotly

Garanta que o gráfico tenha pelo menos os atributos de título e rótulos de dimensão.

````python
#### solução com matplotlib
import numpy as np

# Copiando os dados originais
gols_jitter = wc[['gols_1', 'gols_2']].copy().astype(float)

# Criando um ruído aleatório uniforme entre -0.2 e 0.2 para aplicar o jitter correto (somando)
ruido = np.random.uniform(-0.2, 0.2, size=gols_jitter.shape)
gols_jitter = gols_jitter + ruido

plt.figure(figsize=(8, 6))
plt.scatter(gols_jitter['gols_1'], gols_jitter['gols_2'], alpha=0.5, edgecolors='none')

plt.title('Relação entre Gols do Time Mandante e Visitante (com Jitter)', fontsize=12)
plt.xlabel('Gols do Time Mandante (gols_1)')
plt.ylabel('Gols do Time Visitante (gols_2)')

plt.show()
````

````python
#### solução com seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=gols_jitter, x='gols_1', y='gols_2', alpha=0.5)

plt.title('Relação entre Gols do Time Mandante e Visitante (com Jitter)', fontsize=12)
plt.xlabel('Gols do Time Mandante (gols_1)')
plt.ylabel('Gols do Time Visitante (gols_2)')

plt.show()
````

````python
#### solução com plotly
fig = px.scatter(
    gols_jitter, 
    x='gols_1', 
    y='gols_2', 
    opacity=0.5,
    title='Relação entre Gols do Time Mandante e Visitante (com Jitter)',
    labels={'gols_1': 'Gols do Time Mandante (gols_1)', 'gols_2': 'Gols do Time Visitante (gols_2)'}
)

fig.show()
````

#### Q4.

Apresente um gráfico de barras com o top 10 países que mais participaram de copas do mundo, onde no eixo x devem estar o nome dos países e no eixo y a contagem de participações. Você deve separar a contagem de participações em copas femininas e masculinas, empilhando as barras de cada informação.

No exemplo de barras empilhadas da galeria do matplotlib, imagine que a parte azul são as participações do país em copas masculinas, e em laranja as participações femininas:
* https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_stacked.html

Você deve implementar essa visualização nas três bibliotecas que vimos:
* Matplotlib
* Seaborn
* Plotly

Garanta que o gráfico tenha pelo menos os atributos:
* título
* rótulos de dimensão.
* legenda

````python
# 1. Isolar os países das colunas de mandantes e visitantes
t1 = wc[['time_1', 'ano', 'copa']].rename(columns={'time_1': 'país'})
t2 = wc[['time_2', 'ano', 'copa']].rename(columns={'time_2': 'país'})

# 2. Remover duplicatas por ano/copa para ter participações únicas
part_unicas = pd.concat([t1, t2]).drop_duplicates()

# 3. Criar uma tabela dinâmica (pivot) contando as participações por gênero da copa
df_pivot = part_unicas.pivot_table(index='país', columns='copa', values='ano', aggfunc='count', fill_value=0)

# 4. Calcular o total para definir o Top 10 geral
df_pivot['Total'] = df_pivot['Masculina'] + df_pivot['Feminina']
top10_paises = df_pivot.sort_values(by='Total', ascending=False).head(10).reset_index()

#### solução com matplotlib

plt.figure(figsize=(10, 6))

# Criando as barras empilhadas (Masculina na base, Feminina no topo)
plt.bar(top10_paises['país'], top10_paises['Masculina'], label='Masculina', color='#1f77b4')
plt.bar(top10_paises['país'], top10_paises['Feminina'], bottom=top10_paises['Masculina'], label='Feminina', color='#ff7f0e')

# Adicionando os elementos obrigatórios
plt.title('Top 10 Países com Mais Participações em Copas do Mundo', fontsize=14)
plt.xlabel('Países')
plt.ylabel('Contagem de Participações')
plt.xticks(rotation=45)
plt.legend(title='Categoria')

plt.tight_layout()
plt.show()
````

````python
#### solução com seaborn
plt.figure(figsize=(10, 6))

# Barra do total atrás (representando visualmente onde a feminina vai alcançar no topo)
sns.barplot(data=top10_paises, x='país', y='Total', color='#ff7f0e', label='Feminina')

# Barra masculina na frente, cobrindo a parte de baixo
sns.barplot(data=top10_paises, x='país', y='Masculina', color='#1f77b4', label='Masculina')

# Adicionando os elementos obrigatórios
plt.title('Top 10 Países com Mais Participações em Copas do Mundo', fontsize=14)
plt.xlabel('Países')
plt.ylabel('Contagem de Participações')
plt.xticks(rotation=45)
plt.legend(title='Categoria')

plt.tight_layout()
plt.show()
````

````python
#### solução com plotly
# Formatando os dados para o padrão do Plotly 
top10_long = top10_paises.melt(id_vars='país', value_vars=['Masculina', 'Feminina'], 
                               var_name='Categoria', value_name='Participações')

# Criando o gráfico com empilhamento automático
fig = px.bar(
    top10_long, 
    x='país', 
    y='Participações', 
    color='Categoria',
    title='Top 10 Países com Mais Participações em Copas do Mundo',
    labels={'país': 'Países', 'Participações': 'Contagem de Participações'},
    color_discrete_map={'Masculina': '#1f77b4', 'Feminina': '#ff7f0e'}
)

# Força o Plotly a empilhar as barras caso o tema mude o padrão
fig.update_layout(barmode='stack', xaxis_tickangle=-45)

fig.show()
````

#### Q5.

Vamos fazer um compilado com as estatísticas históricas de copas do mundo!

Com a biblioteca de sua preferência você deve criar 4 subplots organizados em um grid de 2 linhas e 2 colunas. Eles devem conter os seguintes gráficos:
* Linha 1, coluna 1: Gráfico de barras com a quantidade de jogos que aconteceram por ano
* Linha 1, coluna 2: Gráfico de área (referências a seguir) com o total de gols por ano, separando as informações de `gols_1` e `gols_2` para distinguir gols em casa e do time visitante.
* Linha 2, coluna 1: Gráfico de área com o total de cartões por ano, separando as informações de cartões amarelos e cartões vermelhos, mas agregando cartões do time 1 ou time 2. Ou seja, uma área com `cartao_amarelo_1 + cartao_amarelo_2` e outra área com `cartao_vermelho_1 + cartao_vermelho_2`.
* Linha 2, coluna 2: Gráfico de barras com o total de gols contra por ano, somando `gols_contra_1` e `gols_contra_2`.

Referências sobre gráfico de área
* Matplotlib: https://matplotlib.org/stable/gallery/lines_bars_and_markers/stackplot_demo.html#sphx-glr-gallery-lines-bars-and-markers-stackplot-demo-py
* Pandas + Matplotlib: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.area.html
* Seaborn: https://seaborn.pydata.org/generated/seaborn.objects.Area.html
* Plotly: https://plotly.com/python/filled-area-plots/

````python
import pandas as pd
import matplotlib.pyplot as plt

# --- ETAPA DE PREPARAÇÃO DOS DADOS POR ANO ---

# Função auxiliar para contagem de cartões (caso precise redefinir)
def contar_elementos(valor):
    if pd.isna(valor) or str(valor).strip("[]' ") == "":
        return 0
    return len(str(valor).strip("[]").split(','))

# Aplicando as contagens de cartões e gols contra no DataFrame principal
wc['num_amarelos_1'] = wc['cartao_amarelo_1'].apply(contar_elementos)
wc['num_amarelos_2'] = wc['cartao_amarelo_2'].apply(contar_elementos)
wc['num_vermelhos_1'] = wc['cartao_vermelho_1'].apply(contar_elementos)
wc['num_vermelhos_2'] = wc['cartao_vermelho_2'].apply(contar_elementos)

# Tratando os gols contra: preencher nulos com 0 antes de somar
wc['gols_1_contra_num'] = wc['gols_1_contra'].apply(contar_elementos)
wc['gols_2_contra_num'] = wc['gols_2_contra'].apply(contar_elementos)

# Agrupando tudo por ano
dados_ano = wc.groupby('ano').agg(
    qtd_jogos=('ano', 'count'),
    total_gols_1=('gols_1', 'sum'),
    total_gols_2=('gols_2', 'sum'),
    total_amarelos_1=('num_amarelos_1', 'sum'),
    total_amarelos_2=('num_amarelos_2', 'sum'),
    total_vermelhos_1=('num_vermelhos_1', 'sum'),
    total_vermelhos_2=('num_vermelhos_2', 'sum'),
    total_gols_contra_1=('gols_1_contra_num', 'sum'),
    total_gols_contra_2=('gols_2_contra_num', 'sum')
).reset_index()

# Criando colunas consolidadas combinando os dois times
dados_ano['total_amarelos'] = dados_ano['total_amarelos_1'] + dados_ano['total_amarelos_2']
dados_ano['total_vermelhos'] = dados_ano['total_vermelhos_1'] + dados_ano['total_vermelhos_2']
dados_ano['total_gols_contra'] = dados_ano['total_gols_contra_1'] + dados_ano['total_gols_contra_2']


# --- ETAPA DE CONSTRUÇÃO DO GRID DE SUBPLOTS (2x2) ---

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Compilado de Estatísticas Históricas de Copas do Mundo', fontsize=18, fontweight='bold', y=0.96)

# 1. Linha 1, Coluna 1: Gráfico de barras com a quantidade de jogos por ano
axes[0, 0].bar(dados_ano['ano'], dados_ano['qtd_jogos'], color='steelblue', width=2.5, edgecolor='black', alpha=0.8)
axes[0, 0].set_title('Quantidade de Jogos por Ano', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('Ano')
axes[0, 0].set_ylabel('Total de Partidas')

# 2. Linha 1, Coluna 2: Gráfico de área com o total de gols (gols_1 vs gols_2)
axes[0, 1].stackplot(dados_ano['ano'], dados_ano['total_gols_1'], dados_ano['total_gols_2'], 
                     labels=['Gols em Casa (gols_1)', 'Gols Visitante (gols_2)'], colors=['#4CAF50', '#8BC34A'], alpha=0.7)
axes[0, 1].set_title('Evolução do Total de Gols Marcados por Ano', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Ano')
axes[0, 1].set_ylabel('Quantidade de Gols')
axes[0, 1].legend(loc='upper left')

# 3. Linha 2, Coluna 1: Gráfico de área com o total de cartões (amarelos vs vermelhos)
axes[1, 0].stackplot(dados_ano['ano'], dados_ano['total_amarelos'], dados_ano['total_vermelhos'], 
                     labels=['Cartões Amarelos', 'Cartões Vermelhos'], colors=['#FFEB3B', '#F44336'], alpha=0.7)
axes[1, 0].set_title('Total de Cartões por Ano (Agregado Time 1 + Time 2)', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Ano')
axes[1, 0].set_ylabel('Quantidade de Cartões')
axes[1, 0].legend(loc='upper left')

# 4. Linha 2, Coluna 2: Gráfico de barras com o total de gols contra por ano
axes[1, 1].bar(dados_ano['ano'], dados_ano['total_gols_contra'], color='indianred', width=2.5, edgecolor='black', alpha=0.8)
axes[1, 1].set_title('Total de Gols Contra por Ano (gols_contra_1 + gols_contra_2)', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Ano')
axes[1, 1].set_ylabel('Quantidade de Gols Contra')

# Ajustando layout para evitar sobreposições de textos e rótulos
plt.tight_layout(rect=[0, 0.03, 1, 0.93])
plt.show()
````
